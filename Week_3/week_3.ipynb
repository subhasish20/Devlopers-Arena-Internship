### Project: Analyze a Simple Sales Dataset to find: total sales, best-selling product, and create a basic report

```python
# Total Sales = Units Sold Ã— Price per Unit
df["Total_Sales"] = df["Units_Sold"] * df["Price_per_Unit"]
```

```python
# final dataframe 
df.head()
```

```python
# total sales 
total_sales = df["Total_Sales"].sum()
```

```python
total_sales
```

```python
# BEST SELLING PRODUCT (Based on Units Sold)

best_selling = df.groupby("Product")["Units_Sold"].sum().idxmax()
```

```python
best_selling
```

```python
report = f"""
======================
   SALES REPORT
======================

Total Sales: {total_sales}

Best Selling Product: {best_selling}

Average Units Sold: {df["Units_Sold"].mean()}
Highest Units Sold: {df["Units_Sold"].max()}
Lowest Units Sold: {df["Units_Sold"].min()}

----------------------
Report Generated 
"""
```

```python
print(report)
```

```python
# Save report to file
with open("sales_report.txt", "w") as f:
    f.write(report)
```
